```python
class MLStripper(HTMLParser):

	def __init__(self):
		self.reset()
		self.strict = False
		self.convert_charrefs = True
		self.fed = []

	def handle_data(self, d):
		self.fed.append(d)

	def get_data(self):
		return ''.join(self.fed)
```

```python
def strip_tags(self, html):

		s = MLStripper()
		s.feed(html)
		return s.get_data()
```
